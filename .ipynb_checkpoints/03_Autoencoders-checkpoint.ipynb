```python
def fetch_dataset(attrs_name = "lfw_attributes.txt",
                      images_name = "lfw-deepfunneled",
                      dx=80,dy=80,
                      dimx=64,dimy=64
    ):

    #download if not exists
    if not os.path.exists(images_name):
        print("images not found, donwloading...")
        os.system("wget http://vis-www.cs.umass.edu/lfw/lfw-deepfunneled.tgz -O tmp.tgz")
        print("extracting...")
        os.system("tar xvzf tmp.tgz && rm tmp.tgz")
        print("done")
        assert os.path.exists(images_name)

    if not os.path.exists(attrs_name):
        print("attributes not found, downloading...")
        os.system("wget http://www.cs.columbia.edu/CAVE/databases/pubfig/download/%s" % attrs_name)
        print("done")

    #read attrs
    df_attrs = pd.read_csv("lfw_attributes.txt",sep='\t',skiprows=1,) 
    df_attrs = pd.DataFrame(df_attrs.iloc[:,:-1].values, columns = df_attrs.columns[1:])


    #read photos
    photo_ids = []
    for dirpath, dirnames, filenames in os.walk(images_name):
        for fname in filenames:
            if fname.endswith(".jpg"):
                fpath = os.path.join(dirpath,fname)
                photo_id = fname[:-4].replace('_',' ').split()
                person_id = ' '.join(photo_id[:-1])
                photo_number = int(photo_id[-1])
                photo_ids.append({'person':person_id,'imagenum':photo_number,'photo_path':fpath})

    photo_ids = pd.DataFrame(photo_ids)
    df = pd.merge(df_attrs,photo_ids,on=('person','imagenum'))

    assert len(df)==len(df_attrs),"lost some data when merging dataframes"

    all_photos =df['photo_path'].apply(skimage.io.imread)\
                                .apply(lambda img:img[dy:-dy,dx:-dx])\
                                .apply(lambda img: resize(img,[dimx,dimy]))

    all_photos = np.stack(all_photos.values)#.astype('uint8')
    all_attrs = df.drop(["photo_path","person","imagenum"],axis=1)
    
    return all_photos, all_attrs
```

## 1.2. Архитектура модели (1.5 балла)

```python
dim_code = 64
```

Реализуем autoencoder.

```python
class Autoencoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.dim_code = dim_code
        self.encoder = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=256, kernel_size=7, stride=1, padding=3),
            nn.BatchNorm2d(num_features=256),
            nn.ReLU(),
            nn.Conv2d(in_channels=256, out_channels=128, kernel_size=7, stride=1, padding=3),
            nn.BatchNorm2d(num_features=128),
            nn.ReLU(),
            nn.Conv2d(in_channels=128, out_channels=self.dim_code, kernel_size=7, stride=1, padding=3),
            nn.BatchNorm2d(num_features=self.dim_code),
        )
      
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(in_channels=self.dim_code, out_channels=128, kernel_size=7, stride=1, padding=3),
            nn.BatchNorm2d(num_features=128),
            nn.ReLU(),
            nn.ConvTranspose2d(in_channels=128, out_channels=256, kernel_size=7, stride=1, padding=3),
            nn.BatchNorm2d(num_features=256),
            nn.ReLU(),
            nn.ConvTranspose2d(in_channels=256, out_channels=3, kernel_size=7, stride=1, padding=3),
        )
              
    def forward(self, x):
        latent_code = self.encoder(x)
        reconstruction = self.decoder(latent_code)
        return reconstruction, latent_code
```

```python
criterion = F.mse_loss #<loss>

model = Autoencoder()

optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
```

## 1.3 Обучение

```python
from IPython.display import clear_output
from tqdm.notebook import tqdm
n_epochs = 50
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'

model = model.to(device)
train_losses = []
val_losses = []
targets = torch.Tensor()
latent_codes = torch.Tensor()
predicts = torch.Tensor()

for epoch in tqdm(range(n_epochs)):
    model.train()
    train_losses_per_epoch = []
    for X_batch in train_loader:
        optimizer.zero_grad()
        reconstructed, latent_code = model(X_batch.to(device))
        loss = criterion(reconstructed, X_batch.to(device))
        loss.backward()
        optimizer.step()
        train_losses_per_epoch.append(loss.item())
    train_losses.append(np.mean(train_losses_per_epoch))
    model.eval()
    val_losses_per_epoch = []
    with torch.no_grad():
        for X_batch in val_loader:
            reconstructed, latent_code = model(X_batch.to(device))
            loss = criterion(reconstructed, X_batch.to(device))
            val_losses_per_epoch.append(loss.item())
            if epoch == n_epochs - 1:
                targets = torch.cat((targets, X_batch), 0)
                latent_codes = torch.cat((latent_codes, latent_code.cpu().detach()), 0)
                predicts = torch.cat((predicts, reconstructed.cpu().detach()), 0)
            reconstructed, latent_code = model(X_batch.to(device))
            loss = criterion(reconstructed, X_batch.to(device))
            val_losses_per_epoch.append(loss.item())
    val_losses.append(np.mean(val_losses_per_epoch))    
    plt.figure(figsize=(18, 6))
    clear_output(wait=True)
    n = len(X_batch)
    for i in range(n):
        plt.subplot(2, n, i+1)
        plt.imshow(np.rollaxis(X_batch[i].cpu().detach().numpy(), 0, 3))
        plt.title('Input')        
        plt.axis('off')
        plt.subplot(2, n, i+n+1)
        plt.imshow(np.rollaxis(reconstructed[i].cpu().detach().numpy(), 0, 3))
        plt.title('Output')
        plt.axis('off')
    plt.suptitle(
        f'Epoch {epoch+1}/{n_epochs}, loss={round(loss.item(), 5)}'
    )
    plt.show();
```

## 2.1 Архитектура модели и обучение

```python
class VAE(nn.Module):
    def __init__(self):
        super(VAE, self).__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(num_features = 32, momentum = 0.1),
            nn.LeakyReLU(),
            nn.Conv2d(in_channels=32, out_channels=32*2, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(num_features = 32*2, momentum = 0.1),
            nn.LeakyReLU(),
            nn.Conv2d(in_channels=32*2, out_channels=32*4, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(num_features = 32*4, momentum = 0.1),
            nn.LeakyReLU(),
            nn.Conv2d(in_channels=32*4, out_channels=32*8, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(num_features = 32*8, momentum = 0.1),
            nn.LeakyReLU(),
            nn.Conv2d(in_channels=32*8, out_channels=32*16, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(num_features = 32*16, momentum = 0.1),
            nn.LeakyReLU())
        self.mu = nn.Linear(32*16, 4)
        self.logsigma = nn.Linear(32*16, 4)
        self.decoder_input = nn.Linear(4, 32*16)
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(in_channels=32*16, out_channels=32*8, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.BatchNorm2d(num_features = 32*8, momentum = 0.1),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(in_channels=32*8, out_channels=32*4, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.BatchNorm2d(num_features = 32*4, momentum = 0.1),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(in_channels=32*4, out_channels=32*2, kernel_size=3, stride=2, padding=1, output_padding=0),
            nn.BatchNorm2d(num_features = 32*2, momentum = 0.1),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(in_channels=32*2, out_channels=32, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.BatchNorm2d(num_features = 32, momentum = 0.1),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(in_channels=32, out_channels=1, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.Sigmoid())

    def encode(self, x):
        x = self.encoder(x)
        x = torch.flatten(x, start_dim=1)
        mu = self.mu(x)
        logsigma = self.logsigma(x)
        return mu, logsigma
    
    
    def gaussian_sampler(self, mu, logsigma):
        if self.training:
            return self.sample(mu, logsigma)
        else:
            return mu
    
    def decode(self, z):
        z = self.decoder_input(z)
        z = z.view(-1, 512, 1, 1)
        reconstruction = self.decoder(z)
        return reconstruction

    def forward(self, x):
        mu, logsigma = self.encode(x)
        z = self.sample(mu, logsigma)
        reconstruction = self.decode(z)
        return reconstruction, mu, logsigma

    def sample(self, mu, logsigma):
        std = torch.exp(0.5 * logsigma)
        eps = torch.randn_like(std)
        return eps * std + mu
```

Определим лосс и его компоненты для VAE:

Общий лосс будет выглядеть так:

$$\mathcal{L} = -D_{KL}(q_{\phi}(z|x)||p(z)) + \log p_{\theta}(x|z)$$

Формула для KL-дивергенции:

$$D_{KL} = -\frac{1}{2}\sum_{i=1}^{dimZ}(1+log(\sigma_i^2)-\mu_i^2-\sigma_i^2)$$

В качестве log-likelihood возьмем привычную нам кросс-энтропию.

```python
def KL_divergence(mu, logsigma):
    """
    часть функции потерь, которая отвечает за "близость" латентных представлений разных людей
    """
    loss = -0.5 * torch.sum(1 + logsigma - mu.pow(2) - logsigma.exp())
    return loss

def log_likelihood(x, reconstruction):
    """
    часть функции потерь, которая отвечает за качество реконструкции
    """
    loss = nn.BCELoss(reduction='sum')
    return loss(reconstruction, x)

def loss_vae(x, mu, logsigma, reconstruction):
    return KL_divergence(mu, logsigma) + log_likelihood(x, reconstruction)
```

И обучим модель:

```python
criterion = loss_vae

autoencoder = VAE().to(device)
 
optimizer = torch.optim.AdamW(autoencoder.parameters(), lr=3e-4)
```

```python
def train_vae(model, optimizer, criterion, epochs, train_loader, val_loader):
    train_losses = []
    val_losses = []
    for epoch in range(epochs):
        model.train()
        train_losses_per_epoch = []
        for _, X_batch in enumerate(train_loader):
            optimizer.zero_grad()
            X_batch = X_batch[0].to(device)
            reconstructed, mu, logsigma = model(X_batch)
            loss = criterion(X_batch, mu, logsigma, reconstructed)
            loss.backward()
            optimizer.step()
            train_losses_per_epoch.append(loss.item())
        train_loss_mean = np.mean(train_losses_per_epoch)
        train_losses.append(train_loss_mean)
        model.eval()
        val_losses_per_epoch = []
        with torch.no_grad():
            for X_batch_val in val_loader:
                X_batch_val = X_batch_val[0].to(device)
                reconstructed_val, mu_val, logsigma_val = model(X_batch_val)
                loss_val = criterion(X_batch_val, mu_val, logsigma_val, reconstructed_val)
                val_losses_per_epoch.append(loss_val.item())
        val_loss_mean = np.mean(val_losses_per_epoch)
        val_losses.append(np.mean(val_loss_mean))
        clear_output(wait=True)
        fig = plt.figure(figsize=(12,6))
        gs1 = GridSpec(2, 5, left=0.0001, right=1.0, wspace=0.05)
        ax1 = fig.add_subplot(gs1[:-1, :1])
        ax2 = fig.add_subplot(gs1[:-1, 1])
        ax3 = fig.add_subplot(gs1[:-1, 2])
        ax4 = fig.add_subplot(gs1[:-1, 3])
        ax5 = fig.add_subplot(gs1[:-1, -1])
        ax6 = fig.add_subplot(gs1[-1, :1])
        ax7 = fig.add_subplot(gs1[-1, 1])
        ax8 = fig.add_subplot(gs1[-1, 2])
        ax9 = fig.add_subplot(gs1[-1, 3])
        ax10 = fig.add_subplot(gs1[-1, 4])
        axis =[ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9, ax10]
        for k in range(11):
            if k < 5:
                axis[k].imshow(X_batch_val[k].detach().cpu().squeeze(0), cmap='gray')
                axis[k].set_title(f'Real / {str(k+1)}')
                axis[k].grid(False)
                axis[k].axis('off')
            elif 4 < k < 10:
                axis[k].imshow(reconstructed_val[k-5].detach().cpu().squeeze(0), cmap='gray')
                axis[k].set_title('Reconstruction')
                axis[k].grid(False)
                axis[k].axis('off')
        plt.show()
        X_batch.to('cpu')
        del X_batch
        X_batch_val.to('cpu')
        del X_batch_val
        torch.cuda.empty_cache()
        
    return train_losses, val_losses
```

```python
epochs = 50
train_losses, val_losses = train_vae(model=autoencoder, optimizer=optimizer, criterion=criterion, epochs=epochs, train_loader=train_loader, val_loader=test_loader)
```

## 2.2. Latent Representation

```python
def get_latent_vector(self, x):
        mu, logsigma = self.encode(x)
        z = self.reparameterize(mu, logsigma)
        return z
```

## 2.3. Conditional VAE


Давайте теперь допустим, что мы обучили AE на датасете MNIST и теперь хотим генерировать новые картинки с числами с помощью декодера. 

### Архитектура

По сути, единственное отличие от обычного -- это то, что мы вместе с картинкой в первом слое энкодера и декодера передаем еще информацию о классе картинки. 

То есть, в первый (входной) слой энкодера подается конкатенация картинки и информации о классе (например, вектора из девяти нулей и одной единицы). В первый слой декодера подается конкатенация латентного вектора и информации о классе.

Таким образом, при генерации новой рандомной картинки мы должны будем передать декодеру сконкатенированные латентный вектор и класс картинки.

```python
class CVAE(nn.Module):
    def __init__(self):
        super(CVAE, self).__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(num_features = 32, momentum = 0.1),
            nn.LeakyReLU(),
            nn.Conv2d(in_channels=32, out_channels=32*2, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(num_features = 32*2, momentum = 0.1),
            nn.LeakyReLU(),
            nn.Conv2d(in_channels=32*2, out_channels=32*4, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(num_features = 32*4, momentum = 0.1),
            nn.LeakyReLU(),
            nn.Conv2d(in_channels=32*4, out_channels=32*8, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(num_features = 32*8, momentum = 0.1),
            nn.LeakyReLU(),
            nn.Conv2d(in_channels=32*8, out_channels=32*16, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(num_features = 32*16, momentum = 0.1),
            nn.LeakyReLU())
        self.mu = nn.Linear(32*16, 4)
        self.logsigma = nn.Linear(32*16, 4)
        self.embed_class = nn.Linear(10, 28 * 28)
        self.embed_data = nn.ConvTranspose2d(in_channels=1 + 1, out_channels=1, kernel_size=1, padding=0, output_padding=0)
        self.decoder_input = nn.Linear(4 + 10, 32*16)
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(in_channels=32*16, out_channels=32*8, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.BatchNorm2d(num_features = 32*8, momentum = 0.1),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(in_channels=32*8, out_channels=32*4, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.BatchNorm2d(num_features = 32*4, momentum = 0.1),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(in_channels=32*4, out_channels=32*2, kernel_size=3, stride=2, padding=1, output_padding=0),
            nn.BatchNorm2d(num_features = 32*2, momentum = 0.1),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(in_channels=32*2, out_channels=32, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.BatchNorm2d(num_features = 32, momentum = 0.1),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(in_channels=32, out_channels=1, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.Sigmoid())
        
    def reparameterize(self, mu, logsigma):
        std = torch.exp(0.5 * logsigma)
        eps = torch.randn_like(std)
        sample = eps * std + mu
        return sample

    def encode(self, x, label):
        labels = F.one_hot(label, num_classes=10)
        labels = labels.type_as(torch.FloatTensor()).to(device)
        embedded_class = self.embed_class(labels)
        embedded_class = embedded_class.view(-1, 28, 28).unsqueeze(1)
        x = torch.cat([x, embedded_class], dim=1)
        x = self.embed_data(x)
        x = self.encoder(x)
        x = torch.flatten(x, start_dim=1)
        mu = self.mu(x)
        logsigma = self.logsigma(x)
        return mu, logsigma, labels

    def gaussian_sampler(self, mu, logsigma):
        if self.training:
            samples = self.reparameterize(mu, logsigma)
            return samples
        else:
            return mu
    
    def decode(self, z, label):
        labels = F.one_hot(label, num_classes=10)
        labels = labels.type_as(torch.FloatTensor()).to(device)
        z = torch.cat([z, labels], dim=1)
        z = self.decoder_input(z)
        z = z.view(-1, 512, 1, 1)
        reconstruction = self.decoder(z)
        return reconstruction


    def forward(self, x, label):
        mu, logsigma, labels = self.encode(x, label)
        z = self.reparameterize(mu, logsigma)
        z = torch.cat([z, labels], dim=1)
        z = self.decoder_input(z)
        z = z.view(-1, 512, 1, 1)
        reconstruction = self.decoder(z)
        return reconstruction, mu, logsigma
    
    def get_latent_vector(self, x, label):
        mu, logsigma, labels = self.encode(x, label)
        z = self.reparameterize(mu, logsigma)
        return z
```

```python
criterion = loss_vae

c_autoencoder = CVAE().to(device)
 
optimizer = torch.optim.AdamW(c_autoencoder.parameters(), lr=3e-4)
```

```python
def train_cvae(model, optimizer, criterion, epochs, train_loader, val_loader):
    train_losses = []
    val_losses = []
    for epoch in range(epochs):
        model.train()
        train_losses_per_epoch = []
        for _, X_batch in enumerate(train_loader):
            optimizer.zero_grad()
            images, labels = X_batch
            images = images.to(device)
            labels = labels.to(device)
            reconstructed, mu, logsigma = model(images, labels)
            loss = criterion(images, mu, logsigma, reconstructed)
            loss.backward()
            optimizer.step()
            train_losses_per_epoch.append(loss.item())
        train_loss_mean = np.mean(train_losses_per_epoch)
        train_losses.append(train_loss_mean)
        model.eval()
        val_losses_per_epoch = []
        with torch.no_grad():
            for X_batch_val in val_loader:
                images_val, labels_val = X_batch_val
                images_val = images_val.to(device)
                labels_val = labels_val.to(device)
                reconstructed_val, mu_val, logsigma_val = model(images_val, labels_val)
                loss_val = criterion(images_val, mu_val, logsigma_val, reconstructed_val)
                val_losses_per_epoch.append(loss_val.item())
        val_loss_mean = np.mean(val_losses_per_epoch)
        val_losses.append(np.mean(val_loss_mean))
        clear_output(wait=True)
        fig = plt.figure(figsize=(12,6))
        gs1 = GridSpec(2, 5, left=0.0001, right=1.0, wspace=0.05)
        ax1 = fig.add_subplot(gs1[:-1, :1])
        ax2 = fig.add_subplot(gs1[:-1, 1])
        ax3 = fig.add_subplot(gs1[:-1, 2])
        ax4 = fig.add_subplot(gs1[:-1, 3])
        ax5 = fig.add_subplot(gs1[:-1, -1])
        ax6 = fig.add_subplot(gs1[-1, :1])
        ax7 = fig.add_subplot(gs1[-1, 1])
        ax8 = fig.add_subplot(gs1[-1, 2])
        ax9 = fig.add_subplot(gs1[-1, 3])
        ax10 = fig.add_subplot(gs1[-1, 4])
        axis =[ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9, ax10]
        for k in range(11):
            if k < 5:
                axis[k].imshow(images_val[k].detach().cpu().squeeze(0), cmap='gray')
                axis[k].set_title(f'Real / {str(k+1)}')
                axis[k].grid(False)
                axis[k].axis('off')
            elif 4 < k < 10:
                axis[k].imshow(reconstructed_val[k-5].detach().cpu().squeeze(0), cmap='gray')
                axis[k].set_title('Reconstruction')
                axis[k].grid(False)
                axis[k].axis('off')
        plt.show()
        images.to('cpu')
        del images
        images_val.to('cpu')
        del images_val
        torch.cuda.empty_cache()
        
    return train_losses, val_losses
```

```python
epochs = 50

train_losses, val_losses = train_cvae(model=c_autoencoder, optimizer=optimizer, criterion=criterion, epochs=epochs, train_loader=train_loader, val_loader=test_loader)
```
